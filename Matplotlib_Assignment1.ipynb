```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
plt.figure(figsize=(10,10))

lst = ['USA', 'Saudi Arabia', 'france ', 'Eygpt']
  
# Calling DataFrame constructor on list
data= pd.DataFrame(lst)
x=np.array([90,30,60,33])


plt.subplot(1,2,1)
plt.plot(lst,x);
plt.xlabel('Countries')
plt.ylabel('population in Million')
plt.title('plot chart ')


plt.subplot(1,2,2)
plt.bar(lst,x);
plt.xlabel('Countries')
plt.ylabel('population in Million')
plt.title('bar chart ')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(x, labels= lst, autopct='%1.1f%%',colors =colors, startangle=90, radius=1, explode = [0,0,0,0]);
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(lst, x, alpha =0.4,marker = 'o',s=50) 
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(lst, x, alpha =0.4,marker = 'o',s=50) 
plt.ylabel('population',fontsize=16)
plt.xlabel('Countries',fontsize=16)
plt.title('scatter chart',loc='left',color='r')
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python

a=np.array([1,2,0,4,5])
y=np.array([3,6,9,10,2])
plt.plot(a, c='#90DAC3')
plt.plot(y, c ='lightgreen')
plt.plot([12,13, 10, 20,13], color ='blue')
plt.legend(['A','B','C'] ,  loc =2)
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python

plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 30,c='lightgreen')

plt.subplot(1,3,1) 
plt.plot(a, c='#90DAC3')
plt.title('a',fontsize=20);

plt.subplot(1,3,2) 
plt.plot(y, c='#90DAC3')
plt.title('x',fontsize=20);

plt.subplot(1,3,3) 
plt.plot([12,13, 10, 14,13], c='#90DAC3')
plt.title('y',fontsize=20);
```
